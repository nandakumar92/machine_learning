```python
left_df=new_df.select('video_id','trending_date','title','channel_title')
right_df=new_df.select('video_id','trending_date','category_id','views','likes')
```

```python
join_df=left_df.join(right_df,on=['video_id','trending_date'],how='inner')
```

```python
join_df.show(10)
```

```python
# filtering and grouping 
from pyspark.sql.functions import expr

join_df = join_df.withColumn('views_f', expr('cast(views as int)'))
join_df.groupBy('video_id','trending_date').avg('views_f').show()
```

```python
 from pyspark.sql.functions import min, sum, max

result_df = join_df.groupBy('video_id', 'trending_date').agg(
    min('views_f').alias('min_v'),
    sum('views_f').alias('sum_v'),
    max('views_f').alias('max_v')
)
```

```python
result_df.show(10)
```

```python
result_df.filter(result_df.max_v<1231992).show(10)
```

```python
result_df.filter((result_df.trending_date>'17.6.11') & (result_df.video_id=='NDfQkEURQgY')).show(5)
```

```python
# spark sql

result_df.createOrReplaceTempView('results')
spark.sql("select * from results where video_id='NDfQkEURQgY'").show(10)
```

```python
spark.catalog.listTables()
```

```python
## udfs 
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType

def to_upper(word):
    if word is not None:
        return word.upper()
    
to_upper_udf=udf(to_upper,StringType())
```

```python
result_df.show(5)
```

```python
result_df_upper=result_df.withColumn('upper_video_id', to_upper_udf(result_df['video_id']))
```

```python
result_df_upper.show(5)
```

```python
# udf with spark sql

result_df.createOrReplaceTempView('results_temp')
```

```python
def to_upper(word):
    if word is not None:
        return word.upper()
    
spark.udf.register("to_ucase", to_upper, StringType())

spark.sql("select video_id, to_ucase(video_id) as u_video_id from results_temp").show(10)
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import col, row_number, rank, dense_rank, lag, lead, sum as _sum
```

```python
result_df.show()
```

```python
windowSpec=Window.orderBy('trending_date')
```

```python
result_df.withColumn("row_number", row_number().over(windowSpec)).show()
result_df.withColumn("rank", rank().over(windowSpec)).show()
```

```python
from pyspark.sql.functions import desc 

windowSpec=Window.partitionBy('trending_date').orderBy(desc('sum_v'))
```

```python
result_df.withColumn("rank", rank().over(windowSpec)).orderBy(desc('trending_date')).show()
```
